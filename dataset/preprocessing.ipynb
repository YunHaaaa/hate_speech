```python

import os
import pandas as pd 
import numpy as np
```

```python
dev = pd.read_csv('dev.tsv', delimiter='\t')
train = pd.read_csv('train.tsv', delimiter='\t')
train_ = pd.read_csv('train_.tsv', delimiter='\t')
val = pd.read_csv('val.tsv', delimiter='\t')
val_ = pd.read_csv('val_.tsv', delimiter='\t')
```

```python
train_val = pd.concat([train, val], ignore_index=True)
```

```python
train_val
```

```python
train_val_ = pd.concat([train_, val_], ignore_index=True)
```

```python
# comment와 hate 열 추출
train_val_comment_hate = train_val_[['comment', 'hate']].copy()

# 열 이름 변경 
train_val_comment_hate.columns = ['content', 'label']

# 'none'을 0으로, 'offensive'와 'hate'를 1로 변환
train_val_comment_hate['label'] = train_val_comment_hate['label'].apply(lambda x: 0 if x == 'none' else 1)
```

```python
train_val_comment_hate
```

```python
# dev 데이터프레임에서 필요한 열 추출
dev_comment_hate = dev[['문장', '혐오 여부']].copy()

# 열 이름 변경
dev_comment_hate.columns = ['content', 'label']
```

```python
dev_comment_hate
```

```python
# train_val, train_val_comment_hate, dev_comment_hate 데이터프레임 합치기
combined = pd.concat([train_val, train_val_comment_hate, dev_comment_hate], ignore_index=True)

# null 값이 있는 행 삭제
combined = combined.dropna()

# 중복된 데이터 삭제
combined = combined.drop_duplicates()

# label 열 정수로 변환
combined['label'] = combined['label'].astype(int)
```

```python
combined
```

```python
combined.to_csv('combined_data.csv', index=False)
combined.to_csv('combined_data.tsv', sep='\t', index=False)
```

```python

!pip install emoji

import re
import emoji
from soynlp.normalizer import repeat_normalize

emojis = ''.join(emoji.EMOJI_DATA.keys())
pattern = re.compile(f'[^ .,?!/@$%~％·∼()\x00-\x7Fㄱ-힣{emojis}]+')
url_pattern = re.compile(
    r'https?:\/\/(www\.)?[-a-zA-Z0-9@:%._\+~#=]{1,256}\.[a-zA-Z0-9()]{1,6}\b([-a-zA-Z0-9()@:%_\+.~#?&//=]*)')

def clean(x):
    x = pattern.sub(' ', x)
    x = url_pattern.sub('', x)
    x = x.strip()
    x = repeat_normalize(x, num_repeats=2)
    return x
```

```python
combined['content'] = combined['content'].apply(clean)
```
